```python
import evalml
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import holidays
```

```python
df = pd.read_excel('/content/supermarket_sales - Sheet1 copy.xlsx')
```

```python
us_holidays = holidays.US()
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
df['is_holiday'] = df['Date'].apply(lambda date: 1 if date in us_holidays else 0)
```

```python
df['month'] = df['Date'].dt.month
df['day_of_week'] = df['Date'].dt.dayofweek
df['week_of_year'] = df['Date'].dt.isocalendar().week
```

```python
X = df.drop(columns = ['Date','Quantity'], axis = 1)
y = df['Quantity']
```

```python
X_train, X_test, y_train, y_test = evalml.preprocessing.split_data(X, y, problem_type = 'regression', test_size = 0.2)
```

```python
automl = evalml.AutoMLSearch(X_train = X_train, y_train = y_train, problem_type = 'regression',max_batches=1)
```

```python
automl.search()
```

```python
automl.rankings
```

```python
best_pipeline = automl.best_pipeline
```

```python
best_pipeline.fit(X_train, y_train)
```

```python
best_pipeline.score(X_test, y_test, objectives = ['R2', 'MAE'])
```

```python
def forecast_demand(pipeline, start_date, end_date):
    dates = pd.date_range(start=start_date, end=end_date)
    forecast_df = pd.DataFrame({'Date': dates})

    forecast_df['is_holiday'] = forecast_df['Date'].apply(lambda date: 1 if date in us_holidays else 0)
    forecast_df['month'] = forecast_df['Date'].dt.month
    forecast_df['day_of_week'] = forecast_df['Date'].dt.dayofweek
    forecast_df['week_of_year'] = forecast_df['Date'].dt.isocalendar().week

    X_forecast = forecast_df.drop('Date', axis=1)
    predictions = pipeline.predict(X_forecast)

    forecast_df['Predicted Demand'] = predictions
    return forecast_df[['Date', 'Predicted Demand']]
```

```python
forecasted_demand = forecast_demand(best_pipeline, '2024-01-01', '2024-01-31')
print(forecasted_demand)
```
